```
# Downloading the YOLOv3 Weights
from os import path
import requests

target_path = 'drive/My Drive/EVA4/Session-13/Assignments/Assignment-1'
yolov3_weights_url = 'https://pjreddie.com/media/files/yolov3.weights'

if(not path.exists('drive/My Drive/EVA4/Session-13/Assignments/Assignment-1/yolov3.weights')):
  print('Weights file doesn\'t Exist. Downloading...')
  r = requests.get(yolov3_weights_url, stream = True)
  with open(target_path+"/yolov3.weights", "wb") as file:  
    for block in r.iter_content(chunk_size = 1024): 
      if block:  
        file.write(block)
else:
  print('Weights file already Exists.')
```

```
# Downloading the OpenCV Source Code for YOLOv3 Object Detection
import zipfile

source_code_url = 'https://pysource.com/wp-content/uploads/2019/06/yolo_object_detection.zip'

if(not path.exists('drive/My Drive/EVA4/Session-13/Assignments/Assignment-1/yolo_object_detection.zip')):
  print('Source Code doesn\'t Exist. Downloading...')
  r = requests.get(source_code_url, stream = True)
  with open(target_path+"/yolo_object_detection.zip", "wb") as file:  
    for block in r.iter_content(chunk_size = 1024): 
      if block:  
        file.write(block)
  
  print('Extracting the Source Code ZIP...')
  zip_ref = zipfile.ZipFile("drive/My Drive/EVA4/Session-13/Assignments/Assignment-1/yolo_object_detection.zip", 'r')
  zip_ref.extractall("drive/My Drive/EVA4/Session-13/Assignments/Assignment-1")
  zip_ref.close()
  print('Extraction FINISHED!')

else:
  print('Source Code already Exists.')
```
